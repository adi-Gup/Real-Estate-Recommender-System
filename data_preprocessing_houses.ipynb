```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(columns=['property_id','link'],inplace=True)
```

```python
# rename columns
df.rename(columns={'rate':'price_per_sqft'},inplace=True)
df.head()
```

```python
df['society'].value_counts()
```

```python
import re
df['society'] = df['society'].apply(lambda name: re.sub(r'\d+(\.\d+)?\s?★', '', str(name)).strip()).str.lower()
```

```python
df['society'].value_counts().shape
```

```python
df[df['society'] == 'nan'].shape
```

```python
df['society'] = df['society'].str.replace('nan','independent')
```

```python
df.head()
```

```python
df['price'].value_counts()
```

```python
df = df[df['price'] != 'Price on Request']
```

```python
def lacs_to_crore(x):
  if type(x) == float:
    return x
  else:
    if 'Lac' in x:
      return round(float(x[0])/100,2)
    else:
      return round(float(x[0]),2)
```

```python
df['price'] = df['price'].str.split(' ').apply(lacs_to_crore)
```

```python
df.head()
```

```python
df['price_per_sqft'].isna().sum()
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.strip().str.split(' ').str.get(1).str.strip().str.replace(',','').str.split('/').str.get(0).str.strip().astype(float)
```

```python
df.head()
```

```python
df['bedRoom'].value_counts()
```

```python
df['bedRoom'].isna().sum()
```

```python
df = df[~df['bedRoom'].isnull()]
```

```python
df.shape
```

```python
df['bedRoom'] = df['bedRoom'].str.strip().str.split(' ').str.get(0).str.strip().astype(int)
```

```python
df.head()
```

```python
df['bathroom'].value_counts()
```

```python
df['bathroom'].isna().sum()
```

```python
df['bathroom'] = df['bathroom'].str.strip().str.split(' ').str.get(0).str.strip().astype(int)
```

```python
df['balcony'].value_counts()
```

```python
df['balcony'].isna().sum()
```

```python
df['balcony'] = df['balcony'].str.strip().str.replace('No','0').str.split(' ').str.get(0)
```
